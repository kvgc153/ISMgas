```python
import sys
sys.path.append("../")
from gaussian2D import gaussian2D
from astropy.io import fits
import matplotlib.pyplot as plt
```

## Horseshoe

```python
data = fits.getdata("../data/horseshoe.fits")

## We wil try to remove the deflector from this galaxy
plt.imshow(data[1,:,:], origin = 'lower')
```

```python
dd = gaussian2D(
    data = data[1,:,:]
)
results = dd._fit({
    'gaussian2D':{
        'x_mean' : 120,
        'y_mean' : 120
    }
})
```

```python
results.keys()
```

```python
## The class comes with a _plotResults function which you 
## can use to check if the fits are good 

dd._plotResults(results)
```
