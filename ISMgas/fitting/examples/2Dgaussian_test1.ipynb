```python
import sys
sys.path.append("../")
from gaussian2D import gaussian2D
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
data = fits.getdata("../data/cutout_194.5969_18.8159.fits")

## We wil try to center on the galaxy in this data
plt.imshow(data[0,:,:], origin = 'lower')
```

```python
dd = gaussian2D(
    data = data[0,:,:]
)
results = dd._fit({
    'gaussian2D':{
        'x_mean' : 150,
        'y_mean' : 50
    }
})
```

```python
results.keys()
```

```python
## The class comes with a _plotResults function which you 
## can use to check if the fits are good 

dd._plotResults(results)
```
