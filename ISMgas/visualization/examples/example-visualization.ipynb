## Lets create some mock data

```python
from ISMgas.visualization.fits import mockData2D 
Z = mockData2D()
```

## Using ISMgas.visualization.fits to scale the image

```python
from ISMgas.visualization.fits import ScaleImage

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
s = ScaleImage(
    image = Z,
    scale= 'percentile', 
    scale_kwargs = {'percentile':95}, 
    cmap = 'viridis'
)

s.plot()
plt.colorbar()
plt.title("Percentile scaling- 95")


plt.subplot(1,2,2)
s = ScaleImage(
    image = Z,
    scale= 'percentile', 
    scale_kwargs = {'percentile':99.99}, 
    cmap = 'viridis'
)

s.plot()
plt.colorbar()
plt.title("Percentile scaling- 99.99")
```
